# DATA PREPROCESSING

### covert date column string->date

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

### Fill missing values
- city

```python
df['City'] = df['City'].fillna('Unknown')
```

- Super over have na where there is no results.
- so replace NA with Draw

```python
df['SuperOver'] = df['SuperOver'].fillna('NoResults')
```

```python
df['SuperOver'].unique()
```

- Winnig team

```python
df[df['WinningTeam'].isna()]
```

```python
df['WinningTeam'] = df['WinningTeam'].fillna('NoResults')
```

- Marging is null only whene if it is Super over or Draw
- so to keep it it simple we replce it by 0

```python
df['Margin'] = df['Margin'].fillna(0)
```

- method column represents D//L (DLS) if rain was there else None
- Replace NA with a meaningful placeholder like "Standard" mean no rain

```python
df[df['method'].notnull()]
```

```python
df['method'] = df['method'].fillna('Standard')
```
